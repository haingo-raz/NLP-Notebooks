# Dependency parsing

Dependency parsing analyzes the grammatical structure of a sentence and determine how words relate to each other. It helps identify key entities, relationships and actions in a text. 

```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load('en_core_web_sm')

text = "Your product is not good. I will never buy it again."

doc = nlp(text)

displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```

```python
text = "Andy and Tom sat on the porch, he was holding his tears."

doc = nlp(text)

displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```

```python
for token in doc:
    print(token.text, token.dep_, token.head.text, token.head.pos_,
          [child for child in token.children])
```
