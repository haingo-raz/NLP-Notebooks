```python
import re
```

```python
text = """Learning is a journey that never ends. One has to be a lifelong learner to succeed in life. 
In fact, the more we learn, the more we realize how little we know and how much there is to learn"""
```

NB:

The caret (^) is used to match the start of a line. When placed at the beginning of a regex pattern, it ensures that the pattern is matched only if it appears at the start of a line.

the dollar sign ($) is used to match the end of a line. When placed at the end of a regex pattern, it ensures that the pattern is matched only if it appears at the end of a line.

```python
result = re.findall(r"^Learning$", text)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

A . wildcard symbol matches any single character

```python
result = re.findall(r".a.t", text)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

```python
result = re.findall(r"[ha][on][wd]", text)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

+ == "one or more of previous item"

```python
result = re.findall(r"5+", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

| == "or"

```python
result = re.findall(r"l[o|O]+l", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

* == "0 or more of previous item"

```python
result = re.findall(r"l[o|O]*l", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

\ == skip character

```python
result = re.findall(r"[0-9]\.", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

{ } are used to specify the number of occurrences of a preceding pattern.

1. `{n}`: Matches exactly `n` occurrences of the preceding pattern.
    Example: `a{3}` matches "aaa" but not "aa" or "aaaa".

2. `{n,}`: Matches `n` or more occurrences of the preceding pattern.
    Example: `a{2,}` matches "aa", "aaa", "aaaa", and so on.

3. `{n,m}`: Matches between `n` and `m` occurrences of the preceding pattern (inclusive).
    Example: `a{2,4}` matches "aa", "aaa", and "aaaa", but not "a" or "aaaaa".

4. `{,m}`: Matches up to `m` occurrences of the preceding pattern (inclusive).
    Example: `a{,3}` matches "", "a", "aa", and "aaa", but not "aaaa".

5. `{n}?`, `{n,}?`, `{n,m}?`, `{,m}?`: These are the non-greedy versions of the above quantifiers. They match the minimum number of occurrences necessary to satisfy the pattern.
    Example: `a{2,4}?` matches "aa" in "aaaa".

Note: The curly braces `{}` can also be used to specify the exact number of characters to match in character classes. For example, `[a-z]{3}` matches any three lowercase letters.

```python
result = re.findall(r"[0-9]{5}", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

( ) are used for grouping and capturing.

For example, `(ab)+` matches one or more occurrences of the sequence "ab".

For example, in the pattern `(\d+)-(\d+)`, the first capturing group captures one or more digits before the hyphen, and the second capturing group captures one or more digits after the hyphen.

```python
result = re.findall(r"[0-9]{3}.[0-9]{3}.[0-9]+", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

[ ] == define a character class. A character class matches any one character out of a set of characters.

[^XYZ]   Matches a single character not in the listed set

```python
result = re.findall(r"[^Phone]", )

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

? == the preceding element is optional and can occur 0 or 1 time

```python
result = re.findall(r"l[o|O]?l", text_sample)

if result:
    for res in result:
        print(res)
else:
    print("Match not found")
```

\s: matches any whitespace characters such as space and tab.

\S: matches any non-whitespace characters.

\d: matches any digit character.

\D: matches any non-digit characters.

\w : matches any word character (basically alpha-numeric)

\W: matches any non-word character.

\b: matches any word boundary (this would include spaces, dashes, commas, semi-colons, etc.
